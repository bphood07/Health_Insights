# Literature Review 

Our chosen health outcome from this data was “Mental health not good for >=14 days among adults aged >=18 years.” Mental health is the overall state of one’s mind. It affects, and is affected by, just about every aspect in one’s life. It is said that somebody’s mental health is poor if they have a mental health disorder which affects their daily life. This is typically referred to as mental illness, which refers to a wide range of mental health conditions which can affect mood, thinking, and behavior [1]. This could include depression, anxiety, schizophrenia, eating disorders, and addiction, to name just a few. There is a point where certain mental health concerns and symptoms can be classified as a mental illness, on the basis that they are ongoing symptoms which cause frequent stress and affect your daily functions [1]. According to the Mayo Clinic, there are many signs and symptoms to watch that could indicate a mental illness. These include feeling sad, confused thinking or struggles concentrating, excessive fear, worrying, or guilt, extreme mood changes, withdrawal from typical activities and relationships, substantial tiredness or lack of energy, sleeping problems, delusions, paranoia, hallucinations, inability to cope with daily stresses, trouble with interpersonal relationships, problems with substance use, major changes in eating habits, changes in sex drive, extreme anger, hostility, or violence, and suicidal thoughts. There are plenty of other symptoms which could also be signs of mental problems, so if one feels they have issues causing stress on their daily lives they should speak to a professional. Although these are mental illnesses, the body is a complicated and intertwined system and sometimes poor mental health can manifest itself into poor physical health. Some physical problems which could be signs of mental illness include stomach pain, back pain, headaches, or any other unexplained aches and pains [1]. 

There have been many interesting trends explored in mental health research over the past couple of decades. There are plenty of different factors which can contribute to a person’s likelihood of developing a mental illness, as well as their ability to treat it. One of these is age. According to a ten year study from the years 2008 to 2018, the prevalence of anxiety in U.S. adults differs amongst different age groups. The study split the participants into four different age groups: 18-25, 26-34, 35-49, and 50+. There was an increase in anxiety among all of the age groups below 50, whereas those 50 and above had steady rates. The overall increase among U.S. adults was from 5.12%-6.68%. However, even more striking is that the most significant increase was in the youngest age group, 18-25, with anxiety rates that nearly doubled over this 10 year period, from 7.97% to 14.66% as shown in the graph below.

![picture](https://ars.els-cdn.com/content/image/1-s2.0-S0022395620309250-gr2.jpg)

It was found that this increase was a statistically significant trend, even after adjusting for different sociodemographics [2]. This study was only focused on anxiety, but it is estimated that this type of increase is similar across many of the more common mental health disorders. This all supports the claims that not only are the proportions of adults who struggle with mental health increasing, but they seem to be increasing even more with the younger generations. This is very worrying as anxiety in younger adults tends to manifest into further psychological problems down the line, if they go without treatment [2]. The findings of this study suggest that there must be more access to mental health services, and more outreach done into these at-risk communities. 

The more at-risk communities are not just made up of those in the lower age group, however. Multiple studies confirm that mental illness is of increasing concern in rural areas of the country rather than the more urban areas. The National Council for Behavioral Health found that the demand for mental health services is stronger than it’s ever been, but the services offered in the U.S. are insufficient. Nearly six in ten adults are wanting to seek mental health services, with more and more of them being younger than ever before [3]. These younger skews are arising mostly from those coming from lower income households, which generally tend to be more prevalent in rural communities. The study went on to explain that 42% of the population saw cost and poor insurance coverage as the top barriers for accessing mental health care. In fact, one in four Americans had to choose between getting mental health care or paying for daily necessities [3]. The Rural Health Information Hub also has similar findings about the rural populations in America. Although the prevalence of mental illness is similar between rural and urban residents, the services available to those residents differ greatly. Some of the key forces behind this issue are accessibility, availability, and acceptability. “Rural residents often travel long distances to receive services, are less likely to be insured for mental health services, and are less likely to recognize an illness. Chronic shortages of mental health professionals exist and mental health providers are more likely to practice in urban centers [than in rural areas]. The stigma of needing or receiving mental healthcare and fewer choices of trained professionals who work in rural areas create barriers to care” [4]. 

The COVID-19 pandemic has also contributed to the mental health trends in the United States. The Covid-19 pandemic and the economic recession that followed has had a negative impact on many people’s mental illness and substance use disorders. About 4 in 10 adults in the U.S have reported symptoms of anxiety or depressive disorder. A health tracking poll from KFF Health from July 2020 found that many adults reported specific negative impacts on their mental health including trouble sleeping (36%), eating (32%), increase in alcohol consumption/substance abuse (12%) and worsening chronic conditions (12%) [5].  The CDC posted similar findings. A household pulse survey was initiated in late April of 2020, as places all over the country and the world began shutting down for safety. The data showed an increase in the frequency of symptoms of anxiety and depression from the start of the pandemic through the phases they took data in. They also compared it to their 2019 National Health Interview Survey (NHIS), which had lower frequencies [6]. Both of these studies describe increasing mental illness symptoms due to COVID-19. This is supported even further by a Journal of Medical Internet Research article which analyzed the search queries of people across the nation in March 2020. At the start of the pandemic, there was a major spike in searches on mental health symptoms and dealing with mental illness. This study examined these types of searches across those states which had issued stay-at-home orders and those who had not. It found that once the stay-at-home orders were issued, mental health queries leveled off [7].  This study was only performed over about a week’s worth of search queries in Google, meaning that this is very limited data and does not represent the effects of staying at home for the extended period that people have been since then, but rather their initial reactions to these orders. It could be that once there was some certainty reestablished in people’s day-to-day routines, they experienced less feelings of anxiety, depression, lack of focus, etc.

COVID-19 didn’t have only a negative effect on mental health in America, though. As contradictory as it may seem, this global pandemic which caused hundreds of thousands of deaths and an economic recession actually might have improved the state of mental health services in the U.S. This pandemic forced the world into a much more virtually connected norm, introducing many video telecommunication services which were adopted in the workplace, in schools, and also in the healthcare industry; the latter became known as ‘telehealth’ and allows people to receive medical care without leaving home. This extreme surge in telehealth provides some hope for treatment of mental illness, especially in places where there is less physical access, such as rural communities, and in the younger population consisting of high prevalence of mental illness but also a highly virtually connected population. A Taylor & Francis Online article from 2013 explains that at that time the need for mental health services was great and only expected to grow and intensify, which it has. It focused on web camera-based telehealth and its impact in potentially overcoming the mental health burden in America. In the United States, the prevalence of mental health disorders has been estimated at about 50% of all adults being diagnosed with at least 1 psychological disorder. But it is estimated that the number of people actually in need of services for their struggles with mental health is much higher, since many simply don’t meet the threshold to classify their symptoms as a disorder and many more don’t have access to these medical professionals, especially true in rural areas [8]. Telehealth allows these people to meet with a mental health professional and get some type of intervention to work on their problems without running into the typical accessibility issues. Evidence shows that telehealth performs very well compared to its face-to-face counterpart, and even better in some areas as many have seen patients be more willing to share and dig deeper due to the “online disinhibition effect” [8], meaning it often eliminates the ‘acceptability’ roadblock as well since the negative stigma is greatly reduced. The surge in telehealth seen from the pandemic has made this type of service much more readily available than when the article was published and it is expected that even after the U.S. has emerged from the pandemic, many healthcare facilities will still implement telehealth options. There are limitations, however, such as the lack of control that comes with technology, which make this emerging method one which must be carefully and thoughtfully implemented.

Unfortunately there is no real proven cure for mental illness. There is also still plenty of research to be done before any definitive prevention techniques can be established either. But what is known is that it can be somewhat prevented and usually can be treated through therapy, exercise, and reduction of stress-inducing activities in daily life. Oftentimes mental illness is something which starts small as a product of one’s environment and, if left untreated, can develop into mental illness which has a much stronger influence on their everyday life. There are 3 different classifications of psychiatric prevention, based on when the intervention is introduced during the disease’s lifetime. Many of the causal factors for mental illness are societal disparities which increase risk for mental illness. Primary prevention is implemented prior to any signs of disease and is meant to reduce or eliminate causal risk factors, thus preventing onset and reducing the prevalence of that disease. Secondary prevention occurs after one has the disease but before seeing any severe symptoms. It is meant to reduce the severity of the mental illness and reduce the chances of  long-term manifestation. Tertiary prevention is used once the illness has already manifested in the patient and has the goal of preventing any further harm or risk of harm [9]. This can essentially be viewed as treatment rather than prevention, since the illness is already established. Common treatments for mental illness are therapy and medication when the disorder is severe enough, known as psychotropic medication. These methods are often combined for the best results. Other recommendations include a healthy diet, normal sleep schedule, and regular physical activity.

For years the link between exercise and physical health has been well-known and proven time and time again. But a growing interest in research is the link between exercise and mental health. Physical activity has many health benefits including living a life with greater mental health. Physical activity releases endorphins and serotonin that improve mood, enable better sleep, and reduce stress which causes conditions such as depression and anxiety [10]. An Australian study from 2013 explained the correlation between exercise and a healthy mental state saying, “There is a strong relationship between physical activity and mental health. # Hypothesis Testing

## One-Sample Hypothesis Test

Mental health is a growing issue in the U.S. In recent years the average percent of those adults who suffered from mental illness was estimated to be about 50%. This hypothesis test aims to test whether or not this statistic is true in our data population. This will be a two-tailed test.

The null hypothesis is that the mean proportion of adults who suffer from mental illness is equal to 50%. 

The alternate hypothesis is that the mean proportion of adults who suffer from poor mental health is not equal to 50%.

The hypotheses are as follows:<br>
<center> $H_0:\mu = 0.50$<br>
$H_1:\mu \ne 0.50$</center><br>

Where:
* $H_0$ = the null hypothesis
* $H_1$ = the alternate hypothesis
* $\mu$ = the population mean

The population standard deviation is unknown so we will use the t distribution.

The formula to find the $t$ statistic is:<br>
<center> $t$ = $\frac{\bar{X} - \mu}{s/\sqrt n}$</center>

* $t$ = the test statistic
* $\bar{X}$ = the sample mean
* $\mu$ = the population mean
* $s$ = the sample standard deviation
* $n$ = the sample size



```
# first we can calculate what we know
X_Bar = mean(df$Mental.health.not.good.for...14.days.among.adults.aged...18.years) # sample mean
mu0 = 0.50 # hypothesized value
s = sd(df$Mental.health.not.good.for...14.days.among.adults.aged...18.years) # sample standard deviation
n = nrow(df) # sample size

# then calculate the test statistic
t = (X_Bar - mu0)/(s/sqrt(n))
t # our test statistic
```

```
# then we can compute the critical values at the 0.05 significance level
alpha = 0.05
t.half.alpha = qt(1-alpha/2, df=n-1)
c(-t.half.alpha, t.half.alpha)
```

```
# we can graph this
tmp = rnorm(100000, mean=0, sd=1)
plot(density(tmp), main='Two Tail Test (uses t.half.alpha)')
# add a line on the left tail
abline(v=t.half.alpha, col='red', lwd=3)
# add the shaded polygon to the RIGHT
dens = density(tmp) # store those value
x1 <- min(which(dens$x >= t.half.alpha))
x2 <- max(which(dens$x <= Inf))
with(dens, polygon(x=c(x[c(x1,x1:x2,x2)]), y= c(0, y[x1:x2], 0), col="skyblue"))

# add a line on the left tail
abline(v=-t.half.alpha, col='red', lwd=3)
# add the shaded polygon to the LEFT
dens = density(tmp) # store those value
x1 <- min(which(dens$x >= -Inf))
x2 <- max(which(dens$x <= -t.half.alpha))
with(dens, polygon(x=c(x[c(x1,x1:x2,x2)]), y= c(0, y[x1:x2], 0), col="skyblue"))

# now add our purple line
abline(v=t, lty=2, lwd=5, col='purple')
# for illustrative purposes we will show it at a much lower value in the rejection region
# if we use the t value of 346, the rest of the graph will be illegible
abline(v=4.5, lty=2, lwd=5, col='purple')
```

```
# we can calculate the two-tailed p value to reach the same conclusions
pval = 2 * pt(t, df=n-1, lower.tail = F)
pval # two tailed p value
```

From the calculations and the graph, it is clear that the purple test statistic lies in the rejection region, well outside of the range between the red critical values (-1.96 to 1.96).

From the p value being so small that it displays zero, we can say that the p value is much less than the significance level of 0.05.

Both of these reasonings bring us to the same conclusion. We reject the null hypothesis. Thus we can say with 95% confidence that the mean proportion of adults who struggle with poor mental health is not equal to 50%.

## Two-Sample Hypothesis Test
Our research also shows that those who are physically active should have an overall better mental state. This two-sample test aims to show that those adults who *do not* perform physical activity during their leisure time have worse mental health than adults who *are* physically active during their leisure time.

The null hypothesis is that the mean proportion of mental health issues in adults who don't exercise is less than or equal to the mean proportion of mental health issues in adults who do exercise.

The alternate hypothesis is that the mean proportion of mental health problems in adults who do not exercise is more than that of the adults who do exercise.

The hypotheses are as follows:<br>
<center> $H_0:\pi_1 \le \pi_2$<br>
$H_1:\pi_1 > \pi_2$</center><br>

Where:
* $H_0$ = the null hypothesis
* $H_1$ = the alternate hypothesis
* $\pi_1$ = the mean proportion of poor mental health in adults who don't frequently exercise during leisure time
* $\pi_2$ = the mean proportion of poor mental health in adults who frequently exercise during leisure time

```
# density plot of proportion with no leisure time activity
plot(density(df$No.leisure.time.physical.activity.among.adults.aged...18.years)
, main= "Density plot of proportion with no leisure time")
polygon(density(df$No.leisure.time.physical.activity.among.adults.aged...18.years),col="darkolivegreen4")
```

```
# we must split the data into 2 different samples...
# one for the adults with exercise during their leisure time
# one for those without
# the distribution of proportions of adults who don't exercise during leisure time appears relatively normal
# we will use the mean of this distribution as the threshold proportion
# adults with "no leisure time physical activity" equal to or below the mean will be considered active
# adults with "no leisure time physical activity" above the mean will be considered not active

# we can split this data into 2 datasets...
# based on the mean proportion of no leisure time activity
mean.inactivity = mean(df$No.leisure.time.physical.activity.among.adults.aged...18.years)
PhysActive = df[df$No.leisure.time.physical.activity.among.adults.aged...18.years <= mean.inactivity, ]
NotActive = df[df$No.leisure.time.physical.activity.among.adults.aged...18.years > mean.inactivity, ]
```

We will state that it is unreasonable to assume the population standard deviations are equal. The formula for the degrees of freedom for unequal standard deviations is:<br>
<center> $df$ = $\frac{[(\frac{s_1^2}{n_1}) + (\frac{s_2^2}{n_2})]^2}{\frac{(\frac{s_1^2}{n_1})^2}{n_1 - 1} + \frac{(\frac{s_2^2}{n_2})^2}{n_2 - 1}}$</center><br>

* $df$ = the degrees of freedom
* $s_1$ = the standard deviation of sample 1
* $s_2$ = the standard deviation of sample 2
* $n_1$ = the size of sample 1
* $n_2$ = the size of sample 2

**NOTE:** If this formula results in a decimal, it is rounded *down* to the nearest integer.

```
# first, we identify all the variables we know
# sample 1 will be those adults who aren't physically active
# sample 2 will be those adults who are physically active
s1 = sd(NotActive$Mental.health.not.good.for...14.days.among.adults.aged...18.years) # standard deviation for inactive adults
s2 = sd(PhysActive$Mental.health.not.good.for...14.days.among.adults.aged...18.years) # standard deviation for active adults
n1 = nrow(NotActive) # number of observations of inactive adults
n2 = nrow(PhysActive) # number of observations of active adults
```

```
# then we substitute into the formula for degrees of freedom
# use deg for df (degrees of freedom) so we don't overwrite dataframe df
deg = (((s1^2/n1) + (s2^2/n2))^2)/((((s1^2/n1)^2)/(n1-1)) + ((s2^2/n2)^2/(n2-1)))
deg # this is the degrees of freedom (unrounded)
deg = trunc(deg)
deg # this is the truncated degrees of freedom (an integer)
```

```
# then we use that calculated degrees of freedom to find the critical values
# this is a one-tail test
alpha = 0.05 # using the 0.05 significance level
t.alpha = qt(1-alpha, df = deg)
t.alpha
```

Now we can develop our **decision rule:** If the test statistic lies above the critical value of 1.65, we reject the null hypothesis.

The formula for the test statistic $t$ for unequal population standard deviations is:<br>
<center> $t$ = $\frac{{\bar{X_1}-\bar{X_2}}}{\sqrt{\frac{s^2_1}{n_1} + \frac{s^2_2}{n_2}}}$ <br></center>

* $t$ = the test statistic
* $\bar{X_1}$ = the mean of sample 1
* $\bar{X_2}$ = the mean of sample 2
* $s_1$ = the standard deviation of sample 1
* $s_2$ = the standard deviation of sample 2
* $n_1$ = the size of sample 1
* $n_2$ = the size of sample 2

```
# most of these variables have already been defined above...
# we define what is remaining
Xbar1 = mean(NotActive$Mental.health.not.good.for...14.days.among.adults.aged...18.years)
Xbar2 = mean(PhysActive$Mental.health.not.good.for...14.days.among.adults.aged...18.years)

# then substitute everything into the formula for the test statistic
t = (Xbar1-Xbar2)/sqrt((s1^2/n1) + (s2^2/n2))
t # this is our test statistic
```

```
# we can create a graph to illustrate these metrics
tmp = rnorm(100000, mean=0, sd=1)
plot(density(tmp), main='Comparing Means with Unequal Standard Deviations', xlim=c(-6, 52))
# add a red line for the critical value
abline(v=t.alpha, col='red', lwd=3)
# add the shaded polygon to show rejection region
dens = density(tmp) # stored the values
x1 = min(which(dens$x >= t.alpha))
x2 = max(which(dens$x <= Inf))
with(dens, polygon(x=c(x[c(x1,x1:x2,x2)]), y= c(0, y[x1:x2], 0), col="skyblue"))

# add a dotted purple line for the test statistic
abline(v=t, col='purple', lwd=5, lty=2)
```

The calculations and graph above clearly illustrate that we choose to reject the null hypothesis. The purple test statistic line of 51.18 lies well beyond the red critical value line of 1.65, deep in the light blue rejection region.

```
# we can also compute the p value to confirm these results
# again, this is a one-tail test
pval = pt(t, deg, lower.tail = F)
pval
```

The p value is so small that it outputs zero. This is less than our significance level of 0.05. Again, this means we reject the null hypothesis. It is highly unlikely, in fact nearly impossible, that the null hypothesis is true. Thus, we can say with 95% confidence that the mean proportion of adults with poor mental health is higher for inactive adults than for active adults.

The logistic classification models use the data to try to predict a binary decision of whether or not a particular observation's proportion of adults with poor mental health is considered to be good mental health or poor mental health, based on a threshold.
